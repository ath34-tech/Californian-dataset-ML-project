### 3.Important Attribute Selector

```python
from sklearn.base import TransformerMixin,BaseEstimator
```

```python
class FeatureSelector(BaseEstimator,TransformerMixin):
    def __init__(self,n_features):
        self.n_features=n_features
        self.feature_importances_ = None
        self.selected_features_ = None
        self.indices=None
    def fit(self,X,y):
        randomForest=RandomForestRegressor()
        randomForest.fit(X,y)
        self.feature_importances_=randomForest.feature_importances_
        indices=np.argsort(self.feature_importances_)[-self.n_features:]
        self.indices=indices
        self.selected_features_=X.columns[indices]
        return self

    def transform(self,X):
        return X[self.selected_features_]
```

```python
fs=FeatureSelector(n_features=4)
```

```python
fs.fit(housing.drop("ocean_proximity",axis=1),housing_labels)
```

```python
fs.feature_importances_
fs.indices
```

```python
num_pipeline=Pipeline([
            ('feature_selector', FeatureSelector(n_features=4)) ,
     ('imputer',SimpleImputer(strategy='median')),
      ('scalar',MinMaxScaler()),
])

cat_pipeline=Pipeline([
    ('encoder', OneHotEncoder())
])

full_pipeline = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', cat_pipeline, cat_attribs)
])
```

```python
housing_prepared = full_pipeline.fit_transform(housing, housing_labels)
```

```python
param_grid = {
        'prepare__num__feature_selector__n_features': [2, 3, 4],
    'prepare__num__imputer__strategy': ['mean', 'median'],
    'predict__n_estimators': [50, 100],
    'predict__max_features': ['auto', 'sqrt', 'log2']
}
grid_search_prep = GridSearchCV(prepare_select_and_predict_pipeline, param_grid, cv=3,
                                scoring='neg_mean_squared_error', verbose=2)
```

```python
grid_search_prep.fit(housing, housing_labels)
```

```python
print("Best Parameters:", grid_search_prep.best_params_)
print("Best Score:", np.sqrt(-grid_search_prep.best_score_))
```
